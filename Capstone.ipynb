```python
weight = []

#user intput weights
for i in range(4):
    user = float(input('Enter weight ' + str(i+1) + ": "))
    weight.append(user)
print('Weights:', weight)

total = 0.0

#get average weight
for i in range(len(weight)):
    total = total + weight[i]
    avg = total / len(weight);
print('Average weight:{:.2f}'.format(avg))

max = weight[0]

#get max weight
for i in range(len(weight)):
    if weight[i] > max:
        max = weight[i]
print('Max Weight:{:.2f}'.format(max))

#get user input jersey number and player rating
#print jersey number and rating
d={}
for i in range(5):
    number=int(input("Enter player %s's jersey number:\n" % str(i+1)))
    rating=int(input("Enter player %s's rating:\n" % str(i+1)))

    if number not in d:
        d[number]=rating
    print()

print("ROSTER")
for k,v in sorted(d.items()):
    print("Jersey number: %d, Rating: %d" % (k,v))


#print menu and ask for user input
while True:
    print('\nMENU')
    print('a - Add player')
    print('d - Remove player')
    print('u - Update player rating')
    print('r - Output players above a rating')
    print('o - Output roster')
    print('q - Quit\n')

    user=input('Choose an option:\n')





    user=user.lower()

#if o, print roster
    if user=='o':
        print("ROSTER")
        for k,v in sorted(d.items()):
            print("Jersey number:%d,Rating:%d" % (k,v))

#if a, ask for jersey number and player rating
    elif user=='a':

        number=int(input("Enter a new jersey number:\n"))
        rating=int(input("Enter player's rating:\n"))

        if number not in d:
            d[number]=rating
        else:
            print("\nThe Player is already in the list") #if number already exists
            
#if d, ask for jersey number and delete player
    elif user== 'd':

        number=int(input("\nEnter a jersey number:\n"))

        if number in d:
            del d[number]
        else:
            print("\nThe Player is not in the list") #if number doesn't exist

#if u, ask for jersey number and new rating
    elif user== 'u':

        number=int(input("\nEnter a jersey number:\n"))

        if number in d:
            rating=int(input("\nEnter a new rating for the player:\n"))
            d[number]=rating
        else:
            print("\nThe Player is not in the list") #if number already exists

#if r, ask for min rating and print players greater than min rating
    elif user== 'r':


        rating=int(input("\nEnter a rating:\n"))

        for k,v in sorted(d.items()):

            if v > rating:
                print("Jersey number:%d,Rating:%d" % (k,v))

#if q, break
    elif user=='q':
        break
```
